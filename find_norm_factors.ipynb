```python
import pickle as pkl
import hist
import matplotlib.pyplot as plt
```

```python
filename = 'outputs/pt0-13000.pkl'
with open(filename, 'rb') as f:
    output = pkl.load(f)
    
```

```python
for i in range(6):
    h = output['ptjet_mjet_g_reco'].project('ptreco', 'mreco')[i, :]
    print(h.sum())
    h.underflow = 0
    h.overflow = 0
    h.plot(density = True)
    plt.xlim(0,200)
```

```python
for i in range(6):
    h = output['ptjet_mjet_g_reco'].project('ptreco', 'mreco')[i, hist.rebin(2)]
    htrue = output['response_matrix_g'].project('ptgen', 'mgen')[i, :]
    print(h.sum())
    print(htrue.sum())
    h.underflow = 0
    h.overflow = 0
    ax1 = h.plot(density = False)
    ax2 = htrue.plot(density = False)
    plt.xlim(0,200)

    #
    plt.show()
```
