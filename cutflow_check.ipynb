```python
import pickle
import pandas as pd

# Load the pickle file
with open(filename, 'rb') as f:
    output = pickle.load(f)

# Extract the data and create a structured DataFrame
for era, data in output['cutflow'].items():
    print(f"Era: {era}")
    
    # Check if data is not empty
    if not data:
        print("No data available for this era.")
        continue
    
    # Separate the entries into 'gen' and 'reco' based on the keys
    gen_data = {}
    reco_data = {}
    first_key = next(iter(data))  # Retrieve the first key safely
    for key, value in data[first_key].items():
        if 'gen' in key or 'Gen' in key:
            gen_data[key] = value
        elif 'reco' in key or 'Reco' in key:
            reco_data[key] = value
        else:
            # Include keys without 'gen' or 'reco' in both tables for reference
            gen_data[key] = value
            reco_data[key] = value
    
    # Convert to DataFrames
    gen_df = pd.DataFrame(list(gen_data.items()), columns=['Cut', 'Gen Count'])
    reco_df = pd.DataFrame(list(reco_data.items()), columns=['Cut', 'Reco Count'])
    
    # Merge the two dataframes
    cutflow_df = pd.merge(gen_df, reco_df, on="Cut", how="outer").fillna("")

    # Display the DataFrame in tabular format
    display(cutflow_df)
```

```python
import csv
import re

for era in cf.keys():
    # Open a new CSV file for each era
    with open(prependstr+f'{era}_cut_counts.csv', mode='w', newline='') as file:
        writer = csv.writer(file)
        
        # Write the headers, including the new Ratio column
        writer.writerow(['HT-Bin', 'Cut Name', 'RECO Level Cutflow', 'Gen Level Cutflow', 'RECO/GEN Ratio'])
        
        for ht in sorted_era:
            if ht in cf[era].keys():
                d = cf[era][ht]
                
                # Extract the HT range (e.g., 100to200) using regex
                ht_bin = re.search(r'HT-(\d+to\d+|\d+toInf|Inf)', ht)
                ht_bin_text = ht_bin.group(1) if ht_bin else "Unknown"
                
                # List of cuts with 'all events' at the top
                cuts = [
                    ('All Events', 'all events', 'all events'),  # 'all events' at the top
                    ('Jet cuts', 'oneRecoJet', 'oneGenJet'),
                    ('Lepton Cuts', 'twoReco_leptons', 'twoGen_leptons'),                
                    ('Two electron & both passing pfRelIso_03', 'z_ptcut_reco', 'z_ptcut_gen'),
                    ('Z object pt cut', 'z_ptcut_reco', 'z_ptcut_gen'),
                    ('Z object pt cut', 'z_ptcut_reco', 'z_ptcut_gen'),
                    ('Z object m cut', 'z_mcut_reco', 'z_mcut_gen'),
                    ('Dphi(Z, jet) > 1.57', 'z_jet_dphi_sel_reco', 'z_jet_dphi_sel_gen'),
                    ('pt assymetry cut', 'z_pt_asym_sel_reco', 'z_pt_asym_sel_gen')
                ]
                
                # Write each cut to the CSV with an empty ratio column
                for cut_name, reco_key, gen_key in cuts:
                    writer.writerow([ht_bin_text, cut_name, d[reco_key], d[gen_key], ""])
                
                # Special case for Jet ID and Final counts
                writer.writerow([ht_bin_text, 'JetID', d['jetid'], 0, ""])  # Jet ID applied only to RECO, so GEN gets 0
                
                # Calculate the ratio for Final counts row
                final_reco = d['jetid_seq']
                final_gen = d['z_pt_asym_sel_gen_seq']
                ratio = final_reco / final_gen if final_gen != 0 else "Undefined"
                
                # Write Final counts row with the calculated ratio
                writer.writerow([ht_bin_text, 'Final counts', final_reco, final_gen, ratio])
```
